```python
#Feature engineering
#Hits per game
df = df.withColumn('HpG', df.H / df.G)
df.printSchema()
```

```python
#Feature engineering continued
#Perecent doubles
#df = df.withColumn('2BpH', df.'2B' / df.H) #-->apparently this syntax doesnt work with vars beginning with number
#Percent triples
#df = df.withColumn('3bpH', df.'3B' / df.H) #-->apparently this syntax doesnt work with vars beginning with number
#Percent HR's
df = df.withColumn('HRpH', df.HR / df.H)
#Percent HR's
df = df.withColumn('RBIpH', df.RBI / df.H)
#Percent HR's
df = df.withColumn('RBIpAB', df.RBI / df.AB)
df.printSchema()
```

```python
#Select specific features for analysis
#Add some extra features here to work with more 
#complex vectorization syntax
df = df.select("RBI","G")
```

```python
#Define training and validation set with 80-20 split
(test, train) = df.randomSplit((0.15, 0.85))
print ('training set N = {}, test set N = {}'.format(train.count(),test.count()))
```

```python
#Vectorization
from pyspark.ml.linalg import Vectors, VectorUDT

# make a user defined function (udf)
sqlc.registerFunction("oneElementVec", lambda d: Vectors.dense([d]), returnType=VectorUDT())

# vectorize the data frames
train = train.selectExpr("G", "oneElementVec(RBI) as RBI")
test = test.selectExpr("G", "oneElementVec(RBI) as RBI")

print(test.orderBy(test.G.desc()).limit(10))
```

```python
# rename to make ML engine happy
train = train.withColumnRenamed("G", "label").withColumnRenamed("RBI", "features")
test = test.withColumnRenamed("G", "label").withColumnRenamed("RBI", "features")
```

```python
from pyspark.ml.regression import LinearRegression, LinearRegressionModel

lr = LinearRegression()
lrModel = lr.fit(train)

type(lrModel)
```

```python
predictionsAndLabelsDF = lrModel.transform(test)

print(predictionsAndLabelsDF.orderBy(predictionsAndLabelsDF.label.desc()).take(10))
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
eval = RegressionEvaluator()
print(eval.explainParams())
```

```python
#Lets evaluate on all error matrics and R^2 here just for practice
eval.setMetricName("rmse").evaluate(predictionsAndLabelsDF)
```

```python
eval.setMetricName("mse").evaluate(predictionsAndLabelsDF)
```
